```python
import pandas as pd
```

```python
feb = pd.read_csv("2020-Feb.csv")
```

```python
feb.tail(40)
```

```python
feb = feb.drop(columns=["category_id", "brand", "price","user_session","category_code", "product_id"])
```

```python
feb.tail(20)
```

```python
feb = feb.groupby("event_type")['user_id'].nunique().reset_index()
```

```python
view_to_cart= feb.loc[feb['event_type'] == 'cart']['user_id'].values[0] / feb.loc[feb['event_type'] == 'view']['user_id'].values[0]
view_to_purchase= feb.loc[feb['event_type'] == 'purchase']['user_id'].values[0] / feb.loc[feb['event_type'] == 'view']['user_id'].values[0]
cart_to_purchase= feb.loc[feb['event_type'] == 'purchase']['user_id'].values[0] / feb.loc[feb['event_type'] == 'cart']['user_id'].values[0]

print(f"View to cart conversion rate: {view_to_cart:.2%}")
print(f"View to purchase conversion rate: {view_to_purchase:.2%}")
print(f"Cart to purchase conversion rate: {cart_to_purchase:.2%}")
```

```python
# First, let's reload the raw data to preserve the date column for analysis
feb_raw = pd.read_csv("2020-Feb.csv")
feb_raw['event_time'] = pd.to_datetime(feb_raw['event_time'])
feb_raw['date'] = feb_raw['event_time'].dt.date

# Check the date range
print("Date range in data:")
print(f"Min date: {feb_raw['date'].min()}")
print(f"Max date: {feb_raw['date'].max()}")
print(f"\nEvent types: {feb_raw['event_type'].unique()}")
print(f"\nTotal rows: {len(feb_raw)}")
```

```python
# Calculate daily conversion rates
daily_stats = feb_raw.groupby(['date', 'event_type'])['user_id'].nunique().reset_index()
daily_stats['date'] = daily_stats['date'].astype(str)  # Convert dates to strings for proper indexing
daily_pivot = daily_stats.pivot(index='date', columns='event_type', values='user_id').fillna(0)
daily_pivot.columns.name = None

# Calculate conversion rates for each day
daily_pivot['view_to_cart'] = daily_pivot['cart'] / daily_pivot['view']
daily_pivot['view_to_purchase'] = daily_pivot['purchase'] / daily_pivot['view']
daily_pivot['cart_to_purchase'] = daily_pivot['purchase'] / daily_pivot['cart']

print("Daily conversion rates (sample around Feb 27):")
print(daily_pivot.loc['2020-02-24':'2020-02-29', ['view', 'cart', 'purchase', 'view_to_cart', 'cart_to_purchase', 'view_to_purchase']])
print("\n" + "="*80)
print(f"February 27th specifically:")
print(daily_pivot.loc['2020-02-27', ['view', 'cart', 'purchase', 'view_to_cart', 'cart_to_purchase', 'view_to_purchase']])
```

1. February 27th has a conversion rate 6.3 cart to purchase (other days only around 0.5-0.9)
2. What is more damming the views are 39 288 and purchases 18 707 on that day and cart only 2935 -> only 7.47% 
3. That is not constant with the rest of our data because that would mean 86,43% of buyers didn't use the cart -> again not constant with the rest of our data

# I decided that for future visualizations and calculations (excluding the median witch accounts for skewing and extreme data) I will exclude the date of 27th of February

```python
# Let's investigate Feb 27 in detail
import datetime

feb27_date = datetime.date(2020, 2, 27)
feb26_date = datetime.date(2020, 2, 26)

print("=== Detailed Analysis for February 27th ===\n")

# Count raw events (not unique users) by event type
feb27_raw_counts = feb_raw[feb_raw['date'] == feb27_date].groupby('event_type').size()
print("Total events by type on Feb 27:")
print(feb27_raw_counts)
print()

# Compare with a normal day (Feb 26)
feb26_raw_counts = feb_raw[feb_raw['date'] == feb26_date].groupby('event_type').size()
print("Total events by type on Feb 26 (normal day):")
print(feb26_raw_counts)
print()

# Look at unique users by type
feb27_users = feb_raw[feb_raw['date'] == feb27_date].groupby('event_type')['user_id'].nunique()
print("Unique users by event type on Feb 27:")
print(feb27_users)
print()

# Check if there's something unusual about purchases on Feb 27
feb27_purchases = feb_raw[(feb_raw['date'] == feb27_date) & (feb_raw['event_type'] == 'purchase')]
print(f"Number of purchase events on Feb 27: {len(feb27_purchases)}")
print(f"Unique users with purchases on Feb 27: {feb27_purchases['user_id'].nunique()}")
print()

# Check if users made multiple purchases
feb27_user_purchase_counts = feb27_purchases['user_id'].value_counts()
print(f"Users making multiple purchases on Feb 27:")
print(f"Max purchases by one user: {feb27_user_purchase_counts.max()}")
print(f"Average purchases per user: {feb27_user_purchase_counts.mean():.2f}")
print(f"Users with 2+ purchases: {(feb27_user_purchase_counts > 1).sum()}")
```
