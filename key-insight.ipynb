# Funnel performance
### Median conversion rates:
-	View → Cart: 13.58%
-	View → Purchase: 7.5%
-	Cart → Purchase: 53.62%
-	The largest drop-off occurs between View and Cart, indicating early-stage funnel friction.
### High purchase intent once in cart
-	Cart → Purchase conversion remains strong across all months.
-	October shows the highest cart-to-purchase conversion (~0.96), stabilizing around 0.5 thereafter.
### Seasonality effects
-	Purchases increase from October (0.74M) to December (1.16M), decline in January (0.83M), and rebound in February (1.20M).
-	December shows the strongest conversion performance, followed by a post-holiday drop in January.
### Time to purchase
-	Median time to purchase remains consistently low (~3–4 minutes).
-	January and February exhibit the shortest purchase times, suggesting higher buyer intent.
### Purchase value
-	Median purchase price is stable (~$179) from October to December.
-	Median price increases from January to March (~$181), despite lower overall volume.
### Buyer behaviour insight
Post-holiday shoppers are fewer but more decisive and higher-spending.
High-volume months skew toward faster drop-offs and lower median spend.
## Recommendations
-	Improve View to Cart conversion
-	Optimize product pages, pricing clarity, trust signals, and calls-to-action.
-	** Leverage high-intent post-holiday users **
-	** Focus on upselling, premium products, and loyalty strategies in January–February. **
-	Align strategy with seasonality
